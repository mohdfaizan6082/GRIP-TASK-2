# IMPORTING LIBERARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# IMPORTING DATA

```python
data = pd.read_csv("C:\\Users\mohdf\OneDrive\Desktop\Excel Files\Model.csv")
```

```python
data.head(5)
```

# Preaparing Data

```python
X = data.iloc[:,[0]]                   #independent variable
Y=data.iloc[:,1]                        #dependent variable
```

# Regression Algorithm

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=200)
```

```python
from sklearn.linear_model import LinearRegression
Lr=LinearRegression()
Lr.fit(X_train,Y_train)
print("Successful")
```

# Plotting the distribution of scores

```python
plt.figure(figsize=(10,6))
plt.scatter(X_train,Y_train,c="g",plotnonfinite=True,linewidths=5)
plt.title("Simple Linear Regression",fontsize=20)
plt.xlabel('Hours Studied',fontsize=20)  
plt.ylabel('Percentage Score',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,Y,c="g",plotnonfinite=True,linewidths=5)
plt.plot(X,Lr.predict(X),color="red")
plt.title("Simple Linear Regression   Degree=1",fontsize=25)
plt.xlabel('Hours Studied',fontsize=20)  
plt.ylabel('Percentage Score',fontsize=20)
plt.show()
```

# Making Prediction

```python
y_pred=Lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print("R-Squared Value for Training Set: {:.3f}".format(Lr.score(X_train,y_train)))
print("R-Squared Value for Test Set: {:.3f}".format(Lr.score(X_test,y_test)))
```

```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred))
```
